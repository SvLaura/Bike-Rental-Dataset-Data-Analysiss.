# Data cleaning

## 0. Delete duplicates:

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True)
```

```python
df[df.duplicated()]
```

## 1. Change data types:

```python
# convert to datetime
col = ['start_time','end_time', 'started_at','ended_at']
for i in col:
    df[i]  = pd.to_datetime(df[i])
```

```python
# convert to float
col = ['start_station_latitude', 'start_station_longitude','end_station_latitude','end_station_longitude',
       'start_lat','start_lng','end_lat','end_lng']
for i in col:
    df[i]  = pd.to_numeric(df[i], errors='coerce', downcast='float')
```

```python
# convert to int
col = ['bike_id','start_station_id','end_station_id','member_birth_year','duration_sec']
for i in col:
    df[i]  = pd.to_numeric(df[i], errors='coerce', downcast='integer')
```

```python
df.info()
```

## 1. Missing values:

```python
#check if we can replace Null values in old columns on new coluns values:

old_col = ['start_time','end_time','start_station_latitude','start_station_longitude','end_station_latitude',
          'end_station_longitude','user_type']
new_col = ['started_at','ended_at','start_lat','start_lng','end_lat',
          'end_lng','member_casual']

for i in range(len(old_col)):
    #check `theory: if new_col were added later instead old_col':
    ch = df[~df[old_col[i]].isnull()&~df[new_col[i]].isnull()].shape[0]
    
    #replace Null if theory is true:
    if ch == 0:
        df[old_col[i]].fillna(df[new_col[i]], inplace=True)
    #or error:
    else:
        print('Check columns: ' + old_col[i] + ', ' + new_col[i])    
```

```python
# drop columns (repeated):
df.drop(new_col, axis = 1, inplace = True)
```

```python
# drop column - duration_sec 	:
df.drop(['duration_sec'], axis = 1, inplace = True)
```

```python
# check empty rows in as a % of all data:
100 * df.isnull().sum() / len(df)
```

We can assume that some columns where %of missing value is higher 50% were added later.
We should keep it for better analysis for the years when they were added

```python
df['start_year']  = df['start_time'].dt.year
df['start_month']  = df['start_time'].dt.month
df['start_day']  = df['start_time'].dt.day
df['start_weekday'] = df['start_time'].dt.weekday
```

```python
import numpy as np
df['duration_min'] = round((df['end_time'] - df['start_time'])/np.timedelta64(1, 'm'))
```

```python
# drop columns :
df.drop(['start_time','end_time','start_station_id','end_station_id','bike_id','ride_id'], axis = 1, inplace = True)
```
